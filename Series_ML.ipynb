# Criando Features _ Brent_FOB

##  Mínimo, Máximo, Média e Desvio

```python
def percent_change(series):
    #coleta todos da janela, menos o último
    previous_values=series[:-1]
    last_value=series[-1]
    #Calcula a % entre o último e a média da janela
    percent_change= (last_value-np.mean(previous_values))
    np.mean(previous_values)
    return percent_change
    
    
    
```

```python
Brent_perc=serie_x.Brent_FOB.rolling(20).aggregate(percent_change)
```

```python
Brent_perc.plot()
```

```python
features_to_calculate=[np.min, np.max, np.mean, np.std]
feats_1=Brent_perc.rolling(20).aggregate(features_to_calculate).dropna()
feats_1.columns=['f_min','f_max','f_mean','f_std']
```

```python

feats_1.plot()
plt.legend()
plt.show()
```

```python
feats_1.head()
```

## Percentis

```python
from functools import partial
```

```python
percentiles= [1,10,25,50,75,90,99]

percentile_functions=[partial(np.percentile, q=percentile) for percentile in percentiles]
```

```python
def autocorrel(series,lag):
    
    def percent_change(series):
        #coleta todos da janela, menos o último
        previous_values=series[:-1]
        last_value=series[-1]
        #Calcula a % entre o último e a média da janela
        percent_change= (last_value-np.mean(previous_values));np.mean(previous_values)
        return percent_change


    series_perc=series.rolling(20).aggregate(percent_change)
   
    
    
    shifts=np.arange(1,lag).astype(int)
    
    shifted_data= {"lag_{}".format(day_shift): series_perc.shift(day_shift) for day_shift in shifts}
    series_perc_shift= pd.DataFrame(shifted_data)
    
    X=series_perc_shift.fillna(np.nanmedian(series_perc_shift))
    y=series_perc.fillna(np.nanmedian(series_perc))
    
    model=Ridge()
    model.fit(X,y)
    
    
    
    fig, (ax1, ax2)= plt.subplots(1,2, figsize=(15,5))
    y.plot(ax=ax1)
    ax1.legend()
    ax2.bar(shifts ,model.coef_)
    #ax2.set_xticklabels(label, rotation='vertical', fontsize=16)
    plt.xticks(shifts,X.columns)
    ax2.set(title='Coeficientes')
    plt.tight_layout()
    #visualize_coefficients(model.coef_,X.columns,ax=axs[1])
    return plt.show()
    
    
```
