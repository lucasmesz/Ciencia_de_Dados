# Comparação Brent com WTI

```python
fig, (ax1, ax2)= plt.subplots(1,2, figsize=(15,5))
serie_x.Brent_FOB.plot(ax=ax1)
ax1.legend()
serie_x.WTI_FOB.plot(ax=ax2)
ax2.legend()
plt.show()
```

```python
serie_x.Brent_FOB.corr(serie_x.WTI_FOB)
```

```python
serie_x.Brent_FOB.diff().corr(serie_x.WTI_FOB.diff())
```

```python
fig, (ax1, ax2)=plt.subplots(1,2,  figsize=(15,5))
ax1.scatter(serie_x.Brent_FOB,serie_x.WTI_FOB)
ax1.set(title="série", ylabel=serie_x.columns[0], xlabel=serie_x.columns[1])


ax2.scatter(serie_x.Brent_FOB.diff(),serie_x.WTI_FOB.diff())
ax2.set(title="diff", ylabel=serie_x.columns[0], xlabel=serie_x.columns[1])
```

```python
fig, (ax1, ax2)=plt.subplots(1,2, figsize=(15,5))
serie_x.Brent_FOB.diff().plot(ax=ax1, grid=True)
ax1.legend()
serie_x.WTI_FOB.diff().plot(ax=ax2, grid=True)
ax2.legend()
```
